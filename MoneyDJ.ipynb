```python
#完成版
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re

a1=[]
a3=[]
a5=[]
a6=[]

# i 給定爬蟲爬的頁數
for i in range(1,10):
    web='https://www.moneydj.com/KMDJ/News/NewsRealList.aspx?index1='+str(i)+'&a=MB010000'
    html = requests.get(web)
    sp = BeautifulSoup(html.text, 'html.parser')
    da = sp.select(".forumgrid tr td a")
    for j in da:
        next_web='https://www.moneydj.com'+j['href']
        html1 = requests.get(next_web)
        sp1 = BeautifulSoup(html1.text, 'html.parser')
        try:
            title=sp1.select("#ctl00_ctl00_MainContent_Contents_lbTitle")[0].text
            date=sp1.select('#ctl00_ctl00_MainContent_Contents_lbDate')[0].text

            data4=sp1.select('#highlight article')[0].text
            data5=re.findall('\S+',data4)
            content=""
            for k in range(len(data5)):
                content += data5[k]

            source='MoneyDJ'

            a1.append(title)
            a3.append(date)
            a5.append(content)
            a6.append(source)
            print(title)
        except:
            print(j)

aa= {"標題":a1,"日期時間":a3,"文章內容":a5,"文章來源":a6}
aa_df=pd.DataFrame(aa, columns=["標題","日期時間","文章內容","文章來源"])
aa_df.to_excel("/Users/adam/Desktop/py_output/news_crawl/MoneyDJ.xlsx")

aa_df




```

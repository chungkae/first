```python
MD=pd.read_csv("C:\\Users\\Adam\\Desktop\\output\\freedom.csv",encoding="ANSI")  #讀取檔案
MD
```

```python


ss=MD[MD['標題'].str.contains('\u6d17\u9322|\u98a8\u96aa',regex=True)|MD['文章內容'].str.contains('\u6d17\u9322|\u98a8\u96aa',regex=True)]
qq=MD[MD['標題'].str.contains('\u6d17\u9322',regex=True)]

ss
```

```python
df=qq['文章內容']
allcontent=''.join(df)
import jieba.posseg as pseg
import jieba.analyse

jieba.set_dictionary('C:\\Users\\Adam\\Desktop\\help\\dict.txt.big.txt')   #繁體字典
jieba.analyse.set_stop_words("C:\\Users\\Adam\\Desktop\\help\\stop_words.txt")  #設定停止詞

jieba.load_userdict("C:\\Users\\Adam\\Desktop\\help\\userdict.txt")  #使用者補充字典
new=re.sub("\d|[\s+\.\!\/_,$%^*(+\"\']+|[+-—！，。？、~@#￥%……&*()（）:]+",'',allcontent)

worde = jieba.cut(new,cut_all=False)
cc=' '.join(worde)
cc
```

```python
ww=jieba.analyse.extract_tags(new,10)  #找前十大出現
print('/'.join(ww))
```

```python
#黑色背景
from wordcloud import WordCloud
wc=WordCloud(font_path="C:\\Users\\Adam\\Desktop\\help\\simkai.ttf") #中文字型的路徑
cloudimg=wc.generate(cc)

import matplotlib.pyplot as plt
plt.imshow(cloudimg)
plt.axis("off")
plt.show()


#以下為白色背景

wc2=WordCloud(font_path="C:\\Users\\Adam\\Desktop\\help\\simkai.ttf",
              #背景色
             background_color="white",
              #最大詞彙數
             max_words=300,
             relative_scaling=.5,
              #最大號字體
             max_font_size=55,
             random_state=42)

cloudimg2=wc2.generate(cc)
# cloudimg2.to_file("pjl_cloud4.jpg") #保存圖片


plt.figure
plt.imshow(cloudimg2)
plt.axis("off")
plt.show()
```

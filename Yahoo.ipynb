```python
#完成版
import requests
from bs4 import BeautifulSoup
import pandas as pd
from selenium import webdriver
import time

a1=[]
a2=[]
a3=[]
a4=[]
a5=[]
a6=[]
a7=[]

# 選擇要爬的版
web=['stock','finance','industry']    #['stock','real-estate','economy','industry','money-career','international-finance','blogs/society-watch','blogs/for-young','blogs/trend','world','technology','politics']
# ['politics','world','stock','finance','industry'] 
for s in web:
    chrome_path = "/Users/adam/Desktop/py_tool/chromedriver" #chromedriver.exe執行檔所存在的路徑
    web = webdriver.Chrome(chrome_path)
    web.get('https://tw.news.yahoo.com/'+s)
    web.set_window_position(0,0) #瀏覽器位置
    web.set_window_size(1000,700) #瀏覽器大小
    time.sleep(2)
    for r in range(3):  #選擇捲軸捲的次數
        web.execute_script('window.scrollTo'+str((r*8000, (r+1)*8000)))
        time.sleep(0.5)
    sp = BeautifulSoup(web.page_source, 'html.parser')
    front1 = sp.select(".Mb(5px) a")
    for i in front1:
        www='https://tw.news.yahoo.com'+i['href']
        headers = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML,like Gecko) Chrome/55.0.2883.103 Safari/537.36', 'Connection':'keep-alive'}
        html = requests.get(www,headers=headers)
        sp1 = BeautifulSoup(html.text, 'html.parser')
        
        sec1 = sp1.select("#Col1-1-HeadComponentTitle h1")
        if len(sec1)!=0:
            sec11 = sec1[0].text  #標題
            print(sec11)
        else:
            sec11=''
        sec2 = sp1.select("#Col1-3-ContentCanvas div p")
        sec3=""  #內文
        for j in sec2:
            a=j.text
            sec3 += a
        sec4= sp1.select('time')
        if len(sec4)!=0:
            sec44= sec4[0].text  #日期
        else:
            sec44=''
        sec5= sp1.select('div.author-name')
        if len(sec5)!=0:
            sec55= sec5[0].text  #記者
        else:
            sec55=''
        sec6= sp1.select('.provider-link a')
        if len(sec6)!=0:
            sec66= sec6[0].text  #文章來源
        else:
            sec6=''

        a1.append(sec11)
        a2.append(s)
        a3.append(sec44)
        a4.append(sec55)
        a5.append(sec3)
        a6.append(sec66)
        
    web.close()
    
aa= {"標題":a1,"版別":a2,"日期時間":a3,"作者":a4,"文章內容":a5,"文章來源":a6}

aa_df=pd.DataFrame(aa, columns=["標題","版別","日期時間","作者","文章內容","文章來源"])

aa_df.to_csv("/Users/adam/Desktop/py_output/news_crawl/yahoo.csv")

aa_df

```
